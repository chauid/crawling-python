```python
import time
import pandas as pd
#라이브러리 불러오기
from selenium import webdriver as wb
# 웹브라우저를 제어하기 위한 모듈
from selenium.webdriver.common.by import By
# 선택자를 찾기 위한 모듈
from selenium.webdriver.common.keys import Keys #웹에 값을 입력하기 위한 모듈
from selenium.webdriver.chrome.service import Service
# 사용자가 사용하는 최신버전의 크롬 드라이버와 일치하는 드라이버를 찾아주는 모듈
from webdriver_manager.chrome import ChromeDriverManager
# 최신버전의 크롬 드라이버를 찾고 다운로드 해주는 모듈
from bs4 import BeautifulSoup as bs
from IPython.display import display
```

```python
driver = wb.Chrome(service=Service(ChromeDriverManager().install()))
driver.get("https://www.naver.com")

# 검색어 입력
search_box = driver.find_element(By.ID, "query")
search_box.send_keys("트럼프")
search_box.send_keys(Keys.ENTER)
time.sleep(2)

# 뉴스 버튼 클릭
news_btn = driver.find_element(By.XPATH, "//a[starts-with(@href, '?ssc=tab.news')]")
news_btn.click()

# 스크롤 다운
news_links = []
while True:
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    news_links = driver.find_elements(By.CLASS_NAME, "UpDjg8Q2DzdaIi4sfrjX")
    # press_name = driver.find_elements(By.CLASS_NAME, "qJmKqE_urdEXbsEObHAM")
    if(len(news_links) < 30):
        time.sleep(1)
        continue
    break

soup = bs(driver.page_source, "lxml")
news_links = soup.select("a.UpDjg8Q2DzdaIi4sfrjX")
href_list = [a["href"] for a in news_links]
press_names = soup.select("a.qJmKqE_urdEXbsEObHAM>span")
print(href_list)
print(press_names)

pnl_content_tag = soup.find(id="pnlContent")
print(pnl_content_tag)
```

```python
len(href_list), len(press_names)
```

```python
press_name_list = []
for i in range(len(press_names)):
    press_text = press_names[i].get_text()
    if press_text not in press_name_list:
        press_name_list.append(press_text)
press_name_list
# len(press_name_list)
```

```python
press_href_list = []
for i in range(len(href_list)):
    press_href = str(href_list[i]).split("://")[1].split("/")[0]
    if press_href not in press_href_list:
        press_href_list.append(press_href)
press_href_list
# len(press_href_list)
```

```python
press_name_list = []
press_href_list = []
press_list = []
for key, (name, href) in enumerate(zip(press_names, href_list)):
    # print(f"Press {key + 1}: {name.get_text()} (https://{str(href).split('://')[1].split('/')[0]})")
    if name.get_text() not in [press['press_name'] for press in press_list]:
        press_list.append({
            'press_name': name.get_text(),
            'press_href': f"https://{str(href).split('://')[1].split('/')[0]}"
        })
        press_name_list.append(name.get_text())
        press_href_list.append(f"https://{str(href).split('://')[1].split('/')[0]}")

press_list
```

```python
df_press = pd.DataFrame({
    'press_name': press_name_list,
    'press_href': press_href_list
})
df_press.to_excel('press_list.xlsx', index=False)
```
