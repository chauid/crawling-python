```python
# 라이브러리 불러오기
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
url = 'https://www.melon.com/chart/index.htm'
req.get(url)
# Response [406] : 클라이언트(요청) 오류가 났다!!
# --> 멜론 페이지는 보안상 컴퓨터로 접근하는 것을 막고 있기 때문에 --> 우회접속
```

```python
# 페이지 우회 접속 방법
h = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/138.0.0.0 Safari/537.36'}
```

```python
res = req.get(url, headers = h)
res

# 우회접속 하는 방법
# F12 개발자모드 -> Network 탭으로 이동
# F5 눌러 새로고침 -> 스크롤을 맨 위로 올려 Document 타입의 문서 클릭
# 스크롤을 맨 밑으로 내려서 'User-Agent' 값을 복사
# 복사한 값이 value, 'Uesr-Agent'가 Key값에 올 수 있도록 dict 구성
# get() 메소드의 headers 속성에 대입 
```

```python
# bs 객체화
soup = bs(res.text, 'lxml')
soup
```

### 데이터 수집 실습
- 노래 제목
- 가수 이름(아티스트)

```python
# 노래 제목
song = soup.select('div.ellipsis.rank01')
song[0].text.strip()

len(song)
```

```python
# 가수이름 수집
singer = soup.select('span.checkEllipsis')
singer
len(singer)
```

```python
# 반복문
# 빈 리스트 만들어서 가수 이름, 노래 제목, 순위 수집
# -> 그 과정에서 노래 제목에 있는 앞,뒤 공백 제거

songlst = []
singerlst = []
rank = []
for i in range(100):
    songlst.append(song[i].text.strip())
    singerlst.append(singer[i].text.strip())
    rank.append(i+1)
```

```python
dic_melon = {'노래 제목' : songlst, '아티스트' : singerlst, '순위' : rank}
```

```python
import pandas as pd
melon = pd.DataFrame(dic_melon)
melon
```

```python
melon.set_index('순위', inplace = True)
```
